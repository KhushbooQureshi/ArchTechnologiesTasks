```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
avgAge = round(df['Age'].mean() , 0)
print(avgAge)
df['Cabin']= df['Cabin'].fillna('U')
df['familySize'] = df['SibSp'] + df['Parch'] +1 


```

```python
df['Age'] = df['Age'].fillna(avgAge)
```

```python
df['Age'].head()
df['Cabin'].head()
```

```python
newdf = df[['Pclass' , 'Sex' , 'Age' , 'Cabin' , 'Fare' , 'Ticket' ,'familySize']].copy()
y=df['Survived']
                     
```

```python
#now we have clean data to work with this the dataset our model will use to be trained on 
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le= LabelEncoder()
newdf['Sex'] = le.fit_transform(newdf['Sex']) 
newdf['Cabin'] = le.fit_transform(newdf['Cabin']) 
newdf['Ticket'] = le.fit_transform(newdf['Ticket']) 
```

```python
from sklearn.model_selection import train_test_split 

X= newdf
Y= df['Survived']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2 , random_state= 42)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=1000)
model.fit(X_train , Y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
print("\nAccuracy:\n", accuracy_score(Y_test, y_pred))
print("\nClassification Report:\n", classification_report(Y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, y_pred))
```
